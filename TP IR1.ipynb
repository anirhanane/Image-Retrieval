```python
path='/home/UR/anirhan/ENT-anirhan/TP IR/smallholidays'
path_r='/home/UR/anirhan/ENT-anirhan/TP IR/smallholidays_resized/'
path_f='/home/UR/anirhan/ENT-anirhan/TP IR/img_vector/'
```

```python
for images in os.listdir(path):

    
    img=cv2.imread(path+"/"+images)
    #img=Image.open(path+"/"+images)
    img_res=cv2.resize(img,(16,16))
    #img_res=np.resize(img,(16,16,3))

    print(img_res.shape)
        
    cv2.imwrite(path_r+images,img_res)
```

```python
for images in os.listdir(path_r):

    img=cv2.imread(path_r+"/"+images)

    img_data = np.array(img)
    name_image=images.split(".")
    print(path_f+name_image[0]+'.npy')
    np.save(path_f+name_image[0]+'.npy',img_data.flatten())
    
```
